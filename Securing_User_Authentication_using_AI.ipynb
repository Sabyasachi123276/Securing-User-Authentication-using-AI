In the field of cybersecurity, Artificial Intelligence (AI) is playing a pivotal role in the protection of users' sensitive information, including the credentials they use to access their network accounts and applications in order to prevent abuse, such as identity theft [1].	

In present scenarios, in the case of e-commerce, home banking, and so on, it is very much necessary to prevent threats that target use's digital identity. Identity theft has become a very common practice in the modern Internet of Things (IoT)-based era as unauthorized access has become more probable phenomena [2]. In the cyberspace dimension with its increased attack surface, which is determined by the exponential growth in the connections with the establishment between humans and machines, and between machines and machines, that makes rise of the risk of information leakage [3]. Hence, protecting user accounts is not only a matter of data integrity but also there is a reputational risk associated with the legal responsibilities of the corporate as per the mandate provided by EU's General Data Privacy Regulation (GDPR). GDPR is playing a very important role for protecting confidential and sensitive information from users [4]. 


**Passwords and Other Common Authentication Practices**

Passwords play a very vital role for protecting the digital identities of the users.  Though the choice of alphanumeric passwords can enhance the robustness, the users are often seen to use the same password for multiple accounts and services for the easy remembrance of passwords. These types of practices put users’ digital identity under the threat of attackers [5]. 

Once the password is cracked, the attacker can implement a series of illegal activities, for instance, money laundering through the creation of bank accounts on behalf of the victims, disguised behind the credentials of the users who are not aware of this kind of illicit activity.
The protection of a user's account is not only limited to the simple verification of the correctness
of the entered password and the correspondence of the password to the user account, but also the various account activities that are recorded, such as simultaneous access from IP addresses may belong to different geographical areas, or the use of different devices, for instance, PCs, and smartphones, browsers, and operating systems that are quite unusual. The aim of this kind of monitoring is to detect credential theft by the attackers, who attempt to access user accounts by using earlier violated passwords.

Anomaly detection procedures make use of automated learning algorithms which can learn to discriminate between different suspicious activities depending upon the habits and behavior of the users themselves. It can be possible to replace the same passwords with authentication procedures like a user's biometric credentials, such as their iris, voice, fingerprints, or face.
If we discuss the other common authentication practices, to check the authenticity, the second authentication factors are explored nowadays. For instance, the insertion of temporary passwords that are transmitted as OTP codes via SMS messages to the user's phone number, or via the email address associated with the user's account. 

This second authentication process will not be useful if the user's email account has been hacked or malware has been installed on a smart phone on which SMS messages are read automatically forwards the OTP codes to the attacker. These scenarios are quite evident of the ineffectiveness of second authentication factors for security purposes.
Hence, it is advisable that the user should not keep all of their personal, sensitive information within a single support.

Spotting of the fake logins is also possible if the use of authentication procedures based on security tokens, for instance, passwords, SMS, OTPs, and so on, should at least be integrated with automated anomaly detection procedures.
Among the anomalies related to the management of user accounts can be any of the following scenarios like Brute force access attempts; aimed at identifying the user's password by entering different passwords within a limited period of time; simultaneous access from IP addresses belonging to different geographical areas, the use of devices, software, and operating systems that are uncommon for the users and frequency and typing speed that's incompatible with human operators. 

Here, we have to keep in mind that our target is to choose a predictive approach by analyzing the past data for identifying fraud [6]. We are not choosing reactive approaches like setting an alarm system for fraudulent activities to cross any predetermined threshold and will be automatically blocked.


**Prevention of Fake Account Creation**

The prevention of creating fake accounts is very much important to protect the authentic users from the illegal activities carried out by those accounts. The identification of fake profiles are possible while observing the activation of numerous new accounts being carried out by the same IP address within a very shorter span of time, for instance, less than an hour. This kind of activity can be carried out by a bot. 


**Account Reputation Scores**

The monitoring of user account activity is an essential practice which can comprise of both newly created and existing accounts to prevent malicious activities carried out by existing accounts for being under control of the attacker. It is desirable to be associated with a measure of reputation scoring estimation on the basis of the behavior held by the associated user. The reputation scoring provides us the necessary scopes of identifying attacks conducted in stealthy mode by preventing the risk of undetected attacks. It easily implements it using alarm systems calibration for monitoring the anomalous and noisy peaks of activity.

For estimating the reputation score associated with each user account, we can consider various features:

•	The total count, frequency of user posts published over a period of time.

•	Providing access to the user account via proxy, VPN, or other IP anonymized systems.

•	The use of uncommon user agents (such as scripts) to log in.

•	The user's speed in typing text on the keyboard.

These features are taken for training the algorithms followed by dynamically estimating the reputation score of an individual user. Hence, classification of suspicious activities is possible with the help of AI algorithms. 


**Example of Keystroke Dynamics Detection for User Authentication**

Keystroke dynamics captures the cadence and rhythm of keypress events which are unique for each person. Though we can’t apply AI based algorithms directly on these raw datasets as these datasets are contaminated with the external random factors, for instance, interruptions, error corrections, or the use of keyboard special function keys like Shift or Caps Lock.

For avoiding the random disturbing factors, it is important to transform the corresponding raw data representing user digitization in a dataset of features that correctly represent the user's keyboard dynamics by cleaning up the data. 
While talking about biometric recognition, we always consider distinctive physical elements for reliably and exclusively tracing back to a specific human user, for instance the iris, face, fingerprints, voice, and so on. Behaviors and habits can be used for the identification of patterns which can be reliably associated with the individual user. Keystroke dynamics play an important role among these biometric behaviors for the reliable identification of different subjects.

The dataset is available for download in .csv format at https:// www. cs. cmu. edu/ ~keystroke/ DSL- StrongPasswordData. csv. The dataset consists of 51 subjects, each typing 400 passwords; also among the measures collected are these hold times (represented in the dataset with the label H): Keydown-keydown time (labeled as DD) and Keyup-keydown time (labeled as UD).


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.neural_network import MLPClassifier
pwd_data = pd.read_csv("https://www.cs.cmu.edu/~keystroke/DSL-StrongPasswordData.csv", header = 0)
# Average Keystroke Latency per Subject
DD = [dd for dd in pwd_data.columns if dd.startswith('DD')]
plot = pwd_data[DD]
plot['subject'] = pwd_data['subject'].values
plot = plot.groupby('subject').mean()
plot.iloc[:6].T.plot(figsize=(8, 6), title='Average Keystroke Latency per Subject')
```

After implementing the dataset splitting, we evaluate the performances of different machine learning  classifiers.

```python
data_train, data_test = train_test_split(pwd_data, test_size = 0.2,
random_state=0)
X_train = data_train[pwd_data.columns[2:]]
y_train = data_train['subject']
X_test = data_test[pwd_data.columns[2:]]
y_test = data_test['subject']
# K-Nearest Neighbor Classifier
knc = KNeighborsClassifier()
knc.fit(X_train, y_train)
y_pred = knc.predict(X_test)
knc_accuracy = metrics.accuracy_score(y_test, y_pred)
print('K-Nearest Neighbor Classifier Accuracy:', knc_accuracy)
```

```python
# Support Vector Linear Classifier
svc = svm.SVC(kernel='linear')
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
svc_accuracy = metrics.accuracy_score(y_test, y_pred)
print('Support Vector Linear Classifier Accuracy:', svc_accuracy)
```

```python
# Multi Layer Perceptron Classifier
mlpc = MLPClassifier()
mlpc.fit(X_train,y_train)
y_pred = mlpc.predict(X_test)
mlpc_accuracy = metrics.accuracy_score(y_test, y_pred)
print('Multi Layer Perceptron Classifier Accuracy:', mlpc_accuracy)
```

Why does the MLP classifier show considerably better results in terms of prediction accuracy? The answer lies in the fact that it represents an artificial neural network (ANN). ANNs constitute the fundamental element of deep learning and are at the base of high potential that characterizes deep learning algorithms, allowing, for example, the classification of enormous amounts of data, the performance of face and speech recognition, or beating a world chess champion such as Kasparov.

The limitations of the single perceptron are overcome by multilayer perceptrons.
In fact, an MLP is made up of multiple layers of artificial neurons, each implemented by perceptrons.
An MLP can have three or more layers of fully connected artificial neurons, which, as a whole, constitute a feedforward network. Importantly, an MLP can approximate any continuous mathematical function; we can, thus, add an arbitrary number of hidden layers that amplify its overall predictive power.

Now let’s draw the confusion matrix for the Multi Layer Perceptron.

```python
# Drawing confusion matrix for Multi Layer Perceptron results
from sklearn.metrics import confusion_matrix
labels = list(pwd_data['subject'].unique())
cm = confusion_matrix(y_test, y_pred)
figure = plt.figure()
axes = figure.add_subplot(111)
figure.colorbar(axes.matshow(cm))
axes.set_xticklabels([''] + labels)
axes.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
```

Hence, I started with the KNeighborsClassifier clustering algorithm followed by the support vector machine and the multilayer perceptron (MLP). Using MLP, we achieved the highest degree of accuracy in predictions, more than 90%.	

Besides I also have shown the graphical representation of the average keystroke latency per subject and the confusion matrix of the results using the MLP.

**Principal component analysis**

In formal terms, PCA consists of selecting the hyperplane of space along which the data
(represented by points in the space) are mostly spread out; this translates, in mathematical
terms, into the search for the axis in which the variance assumes the maximum value.

To identify this axis, we will need to calculate the covariance matrix associated with our
data, identifying the largest Eigenvectors within the matrix, which correspond to the axes
associated with the main components. By doing this we can reduce the dimensionality of
the data.
The concept of Eigenvectors (as well as the associated concept of Eigenvalues) derives from
linear algebra and gives the name to the facial recognition technique based on Eigenfaces.

We can calculate the values of the variance, of the covariance, and represent the matrix of
covariances using the Python library NumPy. In the following example, we show a
covariance matrix represented by a NumPy array of lists (representing the vectors) and
finally, we print the covariance matrix using the instruction print(np.cov(X).T) as
shown in the following example:

```python
import numpy as np
X = np.array([
[3, 0.1, -2.4],
[3.1, 0.3, -2.6],
[3.4, 0.2, -1.9],
])
print(np.cov(X).T)
```

To understand the relevance of the Eigenvectors and Eigenvalues for the PCA, we must remember that a vector (such as the Eigenvector) represents an oriented element (characterized by a direction) in the linear space, while a scalar (such as the Eigenvalue) represents a measure of intensity (without direction).

Therefore, to identify the principal components within a covariance matrix, we need to look for the Eigenvectors that correspond to the higher Eigenvalues values. In this case, we can use the NumPy library to carry out our calculations.

The calculation of Eigenvectors and Eigenvalues (if they exist) is reduced to the following NumPy instructions:

```python
import numpy as np
eigenvalues, eigenvectors = np.linalg.eig(np.array([[2, -4], [4, -6]]))
print(eigenvalues)
print(eigenvectors)
```

**Application of the Principal Component Analysis (PCA) Technique for Facial Recognition**

Here, we shall associate each image in our archive with the name corresponding to the person represented in the image. For implementing this, the reduction in dimensionality of the images to the main components have been implemented, followed by identifying the most relevant features for recognition purposes. These main components are termed as Eigenfaces.
Some Eigenfaces are given in the following figure:


Eigenfaces.png

(Image credits: Wikipedia, at https://en.wikipedia.org/wiki/File:Eigenfaces.png)

Each image of our dataset can, therefore, be interpreted as a combination of these Eigenfaces.
The image dataset is available for download at http:// vis- www. cs. umass. edu/ lfw/ lfw- funneled. tgz.
In this example, we shall use the concept of Multi-layer Perceptron for doing the classification on images.
The output of the Multi-layer Perceptron classifier is shown via classification_report(), which indicates the values of the precision, recall, and F1 score metrics.


```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
lfw = fetch_lfw_people(min_faces_per_person=150)
X_data = lfw.data
y_target = lfw.target
names = lfw.target_names
X_train, X_test, y_train, y_test = train_test_split(X_data, y_target,
test_size=0.3)
pca = PCA(n_components=150, whiten=True)
pca.fit(X_train)
pca_train = pca.transform(X_train)
pca_test = pca.transform(X_test)
mlpc = MLPClassifier()
mlpc.fit(pca_train, y_train)
y_pred = mlpc.predict(pca_test)
print(classification_report(y_test, y_pred, target_names=names))
```
